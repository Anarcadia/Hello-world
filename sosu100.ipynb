```python
#課題３
sosu100 = []
for i in range(2, 101):
    sosu = True
    for n in range(2, i):
        if i % n == 0:
            sosu = False
            break
    if sosu:
        sosu100.append(i)

for x in range(1,101):
    x_a = x//15
    x_b = x//3
    x_c = x//5
    if x%15==0:
        print(x,"3 * 5 * ",x_a)
    elif x%3==0:
        if x in sosu100:
            j=(sosu100.index(x))+1
            print(x,":","3 * ",x_b,"---",j,"番目の素数")
        else:
            print(x,":","3 * ",x_b)

    elif x%5==0:
        if x in sosu100:
            j=(sosu100.index(x))+1
            print(x,":","5 * ",x_c,"---",j,"番目の素数")
        else:
            print(x,":","5 * ",x_c)
    else:
        if x in sosu100:
            j=(sosu100.index(x))+1
            print(x,"---",j,"番目の素数")
        else:
            print(x)
```
